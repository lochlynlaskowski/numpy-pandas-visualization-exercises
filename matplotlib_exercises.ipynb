```python
x1= range(0,150)
y1= [math.sqrt(n) for n in x]

x2 = range(0,150)
y2 = [n**3 for n in x]

x3 = [x * 0.1 for x in range(0,150)]
y3 = [2**n for n in x]

x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x]

plt.figure(figsize = (12,10))


plt.plot(x1,y1, label = '$sqrt(x)$')

plt.plot(x2,y2, label = '$x^3$')

plt.plot(x3,y3, label = '$2^x$')

plt.plot(x4,y4, label = '$1/(x+1)$')

plt.grid(True, ls = '--')
plt.ylim(0,20)
plt.xlim(0,20)
plt.legend()
plt.title('Four Functions')
plt.xlabel('$x$')
plt.ylabel('$y$')
```
