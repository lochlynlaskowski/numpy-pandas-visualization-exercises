```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

from env import get_db_url
import pandas as pd


url = get_db_url("mall_customers")
```

```python
sql = '''
SELECT * FROM customers
'''
mall_customers = pd.read_sql(sql, url)
mall_customers
```

```python
sns.boxplot(data = mall_customers, x = 'age', y = 'annual_income')
```

```python
mall_customers.age.max(), mall_customers.age.min()
```

```python
mall_customers['age_group'] = pd.cut(mall_customers.age, [10,20,30,40,50,60,70])
mall_customers
```

```python
sns.boxplot(data = mall_customers, x = 'age_group', y = 'annual_income')
```

```python
sns.boxplot(data = mall_customers, x = 'age_group', y = 'annual_income', hue = 'gender')
```

```python
sns.violinplot(data = mall_customers, x = 'age_group', y = 'annual_income', hue = 'gender', color = 'green')
```

```python
sns.relplot(data = mall_customers, x = 'spending_score', y = 'age', hue = 'gender')
```

```python
sns.lmplot(data = mall_customers, x = 'spending_score', y = 'age', hue = 'gender')
```
