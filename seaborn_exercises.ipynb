```python
# Use the iris database to answer the following quesitons:
iris = sns.load_dataset('iris')
# What does the distribution of petal lengths look like?
sns.displot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width? Use http://guessthecorrelation.com/ as a hint to how we could visually determine if a correlation exists between two numeric columns.
sns.scatterplot(data = iris, x = 'petal_length', y = 'petal_width', hue = 'species')
# this shows a strong positive correlation
```

```python
iris.petal_length.corr(iris.petal_width)
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length? For this, you'll visualize two numeric columns through the lense of a categorical column.
sns.scatterplot(data = iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
# setosa can be identified outside of the other two species, but versicolor and virginica have too much overlap to identify
```

```python
# Which features would be best used to predict species?
sns.pairplot(data = iris, hue = 'species')

# petal_length vs petal_width has the clearest distinction between the species
```

```python
# Load the swiss dataset and read it's documentation. The swiss dataset is available from pydatset rather than seaborn. Create visualizations to answer the following questions:
from pydataset import data
data('swiss', show_doc = True)
swiss = data('swiss')
swiss

```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)

swiss['is_catholic'] = swiss.Catholic >= 60
swiss
```

```python
# Does whether or not a province is Catholic influence fertility?
sns.boxplot(data = swiss, x = 'is_catholic', y = 'Fertility')
# based off the arbitrary cutoff of 60, it seems that a province being Catholic has an effect on fertility
```

```python
# Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.
from env import get_db_url
url = get_db_url('chipotle')
sql = "SELECT * FROM orders"
orders = pd.read_sql(sql, url)
orders.head()
```

```python
orders['item_price'] = orders.item_price.str.strip('$').astype('float')
```

```python
most_popular = orders.groupby('item_name').item_name.count().sort_values(ascending=False).head(4)
most_popular
```

```python
revenue = orders[orders.item_name.isin(most_popular.index)].groupby('item_name').item_price.sum()
revenue
```

```python
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
popular = sns.barplot(data = revenue, x = 'item_price', y = 'item_name')
popular.set_xlabel("Total Revenue")
popular.set_ylabel("")
```
